```python
from matplotlib import pyplot as plt
```

```python
x=[3,4,5]
y=[10,11,12]
plt.plot(x,y)
plt.title("info")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()
```

```python
#style
x=[1,2,3]
y=[4,5,7]

x1=[5,8,10]
y1=[12,16,6]
plt.plot(x,y,"w",label="line one",linewidth=5)
plt.plot(x1,y1,"r",label="line two",linewidth=5)

plt.title("info")
plt.ylabel("Y.axis")
plt.xlabel("x.axis")

plt.legend()
plt.grid(True,color="y")
plt.show()


```
